```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('var33.csv')
data.head(10)
```

```python
# среднее
avg1 = data['qrqmw'].mean()
avg2 = data['dtqui'].mean()
avg3 = data['tbpwm'].mean()
avg1, avg2, avg3
```

```python
# дисперсия
var1 = data['qrqmw'].var()
var2 = data['dtqui'].var()
var3 = data['tbpwm'].var()
var1, var2, var3
```

```python
index = np.arange(100)
plt.bar(index, data['qrqmw'])
plt.title('qrqmw', fontsize=20)
plt.show()
plt.bar(index, data['dtqui'])
plt.title('dtqui', fontsize=20)
plt.show()
plt.bar(index, data['tbpwm'])
plt.title('tbpwm', fontsize=20)
plt.show()
```

```python
df = pd.DataFrame({'tbpwm' : data['tbpwm'],
        'qrqmw' :  data['qrqmw']})
df.plot()
plt.title('qrqmw + tbpwm', fontsize=20)
plt.show()

df = pd.DataFrame({'dtqui' : data['dtqui'],
        'tbpwm' :  data['tbpwm']})
df.plot()
plt.title('tbpwm + dtqui', fontsize=20)
plt.show()

df = pd.DataFrame({'dtqui' : data['dtqui'],
        'qrqmw' :  data['qrqmw']})
df.plot()
plt.title('qrqmw + dtqui', fontsize=20)
plt.show()
```

```python
#удалю выбросы в qrqmw
data[(data['qrqmw']>50)|(data['qrqmw']<0) ]=avg1
```

```python
#удалю выбросы в tbpwm
data[(data['tbpwm']>50)|(data['tbpwm']<0) ]=avg3
```

```python
# среднее
avg1_afterupdate = data['qrqmw'].mean()
avg2_afterupdate = data['dtqui'].mean()
avg3_afterupdate = data['tbpwm'].mean()
avg1_afterupdate, avg2_afterupdate, avg3_afterupdate
```

```python
# дисперсия
var1_afterupdate = data['qrqmw'].var()
var2_afterupdate = data['dtqui'].var()
var3_afterupdate = data['tbpwm'].var()
var1_afterupdate, var2_afterupdate, var3_afterupdate
```

```python
index = np.arange(5)
d = {'avg_afterupdate' : [avg1_afterupdate, avg2_afterupdate, avg3_afterupdate],
        'avg' : [avg1, avg2, avg3]}
df = pd.DataFrame(d)
df.plot(kind='bar')
plt.title('avg differnece', fontsize=20)
plt.show()
d = { 'var_afterupdate' : [var1_afterupdate, var2_afterupdate, var3_afterupdate],
        'var' : [var1, var2, var3]}
df = pd.DataFrame(d)
df.plot(kind='bar')
plt.title('var difference', fontsize=20)
plt.show()
```

```python
data['qrqmwNorm'] = (data['qrqmw'] - avg1_afterupdate)/(var1_afterupdate**0.5)
data['dtquiNorm'] = (data['dtqui'] - avg2_afterupdate)/(var2_afterupdate**0.5)
data['tbpwmNorm'] = (data['tbpwm'] - avg3_afterupdate)/(var3_afterupdate**0.5)
```

```python
index = np.arange(100)
plt.bar(index, data['qrqmwNorm'])
plt.title('qrqmwNorm', fontsize=20)
plt.show()
plt.bar(index, data['dtquiNorm'])
plt.title('dtquiNorm', fontsize=20)
plt.show()
plt.bar(index, data['tbpwmNorm'])
plt.title('tbpwmNorm', fontsize=20)
plt.show()
```
